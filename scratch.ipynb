```python
import pandas as pd
import datetime
```

```python
brands = pd.read_json("fetch_data/brands.json", lines = True)
users = pd.read_json("fetch_data/users.json", lines=True)
receipts = pd.read_json("fetch_data/receipts.json", lines=True)

for x in [brands, users, receipts]:
    display(x.shape)
    display(x.head())
```

```python
def clean_date(date_dict):
    if type(date_dict) != dict:
        return None
    timestamp = date_dict.get('$date', 0)
    return pd.to_datetime(timestamp, unit='ms')

def clean_id(id_dict):
    if type(id_dict) != dict:
        return None
    if '$id' in id_dict:
        return clean_id(id_dict['$id'])
    return id_dict.get('$oid', -1)
```

```python
for table in [brands, users, receipts]:
    table["id"] = table["_id"].apply(clean_id)
    if 'cpg' in table.columns:
        table['cpg_clean'] = table['cpg'].apply(clean_id)
    
    for col in table.columns:
        if 'clean' not in col.lower() and ('date' in col.lower() or col == 'lastLogin'):
            table[f"{col}_clean"] = table[col].apply(clean_date)

    display(table.head())

```

```python


expected_list_keys = set()
for l in receipts['rewardsReceiptItemList']:
    if not isinstance(l, list):
        continue
    for d in l:
        if not isinstance(d, dict):
            continue
        expected_list_keys |= d.keys()

expected_list_keys

receipts.shape
ritems = receipts.explode('rewardsReceiptItemList').reset_index()[['id', 'rewardsReceiptItemList']]
for field in expected_list_keys:
    ritems[field] = ritems['rewardsReceiptItemList'].apply(lambda x: x.get(field) if isinstance(x, dict) else None)

ritems = ritems[ritems['rewardsReceiptItemList'].apply(lambda x: pd.notna(x))]
ritems.shape
```
